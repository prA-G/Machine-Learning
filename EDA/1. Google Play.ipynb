The size column is mentioned as object column due to presence of M and k, In order to make it numerical column we will convert M & k into numerical value.

We know that 1 MB = 1024 KB

````python
df['Size']
````

````python
df['Size'].unique()
````

````python
def sizeProcess(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M", ""))
        res = res * 1024
        return res
    elif str(item)[-1] == "k":
        res = float(str(item).replace("k", ""))
        return res
    else:
        return str(np.nan)
````

````python
sizeProcess("19M")
````

````python
df_copy['Size'] = df_copy['Size'].apply(sizeProcess)
````

````python
df_copy['Size'].unique()
````

````python
df_copy['Size'].dtypes
````

````python
df_copy.Size = df_copy.Size.astype("float")
````

We can make several object column to numerical column in one go.

```python
char_to_remove = ["+", "," , "$"]
cols_to_remove = ["Installs", "Price"]

for char in char_to_remove:
    for col in cols_to_remove:
        df_copy[col] = df_copy[col].str.replace(char, "")
```

Since we already converted object into numerical column , it will throw error. But by above code we can convert multiple column in numerical column in one go.
