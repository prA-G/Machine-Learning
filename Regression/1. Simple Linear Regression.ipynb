### Technique to Find the Best Fit Line in Simple Linear Regression

In **Simple Linear Regression (SLR)**, the goal is to find the **best-fitting straight line** that predicts the dependent variable (Y) from the independent variable (X).

The equation of the line is:

\[
Y = β_0 + β_1X + ε
\]

Where:
- \( β_0 \) → Intercept  
- \( β_1 \) → Slope  
- \( ε \) → Error term  

---

### The Technique Used: **Least Squares Method**

The **Least Squares Method** is used to find the *optimal line*.  
It minimizes the **sum of squared differences** between actual and predicted values of Y.

\[
\text{Minimize } \sum (Y_i - \hat{Y_i})^2
\]

Where:
- \( Y_i \): Actual values  
- \( \hat{Y_i} \): Predicted values from the regression line  

---

###  Steps:

1. **Assume a linear model:**  
   \( \hat{Y} = β_0 + β_1X \)

2. **Compute slope (β₁):**  
   \[
   β_1 = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sum (X_i - \bar{X})^2}
   \]

3. **Compute intercept (β₀):**  
   \[
   β_0 = \bar{Y} - β_1\bar{X}
   \]

4. **Form regression line:**  
   Substitute β₀ and β₁ in \( \hat{Y} = β_0 + β_1X \)

5. **Evaluate fit:**  
   - Use **R² score** or **Residual Plots** to check how well the line fits.

---

### In short:
> The *Least Squares Method* finds the best line by making the total squared prediction errors as small as possible — balancing underfit and overfit.


### Sum of Squared Error (SSE) and Mean Squared Error (MSE)

In regression analysis, both **SSE** and **MSE** measure how far the predicted values are from the actual values.  
They help evaluate the **accuracy and performance** of a regression model.

---

###  1. Sum of Squared Errors (SSE)

**Definition:**  
SSE is the total of all squared differences between actual and predicted values.

\[
SSE = \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2
\]

Where:
- \( Y_i \): Actual values  
- \( \hat{Y_i} \): Predicted values  
- \( n \): Number of data points  

**Interpretation:**  
- Lower SSE → Better model fit.  
- If SSE = 0 → Perfect predictions (rare in real-world data).

---

###  2. Mean Squared Error (MSE)

**Definition:**  
MSE is the **average** of the squared errors.  
It’s simply the SSE divided by the number of observations.

\[
MSE = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2
\]

**Interpretation:**  
- MSE measures the **average squared deviation** of predictions from actual values.  
- Smaller MSE = more accurate model.  
- MSE penalizes **larger errors more heavily** due to squaring.

---

###  Relation Between SSE and MSE:
\[
MSE = \frac{SSE}{n}
\]

---

### Summary:

| Metric | Formula | Purpose |
|:-------|:---------|:---------|
| **SSE** | \( \sum (Y_i - \hat{Y_i})^2 \) | Total error magnitude |
| **MSE** | \( \frac{1}{n}\sum (Y_i - \hat{Y_i})^2 \) | Average error magnitude |
| **Lower Value** | → Better model performance |  |



### Cost Function (CF)

The **Cost Function** measures how well the regression model fits the data.  
It tells us how far the model’s predictions are from the actual values.

In **Linear Regression**, the most common cost function is the **Mean Squared Error (MSE)**:

\[
J(β_0, β_1) = \frac{1}{2n} \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2
\]

Where:
- \( J(β_0, β_1) \): Cost function  
- \( Y_i \): Actual value  
- \( \hat{Y_i} \): Predicted value  
- \( n \): Number of samples  

The goal of training a regression model is to **minimize this cost function** —  
that’s how we find the best \( β_0 \) (intercept) and \( β_1 \) (slope).

---

###  Key Idea:
Lower **CF** → smaller errors → better fit.

This minimization is usually done using **Ordinary Least Squares (OLS)** or **Gradient Descent**.


### Why is CF a penalty?
Because it literally penalizes the model for being wrong.

Every time your prediction ^Yi strays from the actual Yi, the cost function adds a penalty — the square of that error. The farther off you are, the bigger the punishment.

Think of it like a teacher grading an exam: one wrong answer might cost 1 mark, but a completely off-topic answer costs 5. The cost function does the same — bigger errors hurt more.

So:

Small errors → small penalty → low cost.

Big errors → large penalty → high cost.

And when you train the model, you’re just trying to minimize that total penalty, finding the line that annoys the cost function the least.
