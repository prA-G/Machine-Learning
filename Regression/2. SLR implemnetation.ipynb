Now since we have target variables, feature names we can build dataFrame

```python
data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
data.head()
```

```python
data['target'] = diabetes.target
```

```python
data
```

### Data Cleaning

Seperating Data into X and Y

```python
X = data[['bmi']] # one independent variable
y = data['target'] # dependent variable
```

```python
X, y
```

### Train-Test-Split

Split array or matrices in random data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

random_state = 1 when fixed helps us to reproduce the same result again and again for 1 data

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Model Training

**Linear Regression** : The following are a set of methods intended for regressions in which the target value is expected to be a linear combination of the features.

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
X_train
```

```python
y_train
```

Training of Model

```python
model.fit(X_train, y_train)
```

At the backend it will use gradient descent to give optimal coefficients

```python
model.coef_  #attribute of the model
```

```python
model.intercept_  #attribute of the model
```

```python
y_pred = model.predict(X_test)  #predicting for a new value
```

```python
y_pred
```

### Visualization of result

```python
plt.scatter(X_test, y_test, color='purple', label='Actual Data')
plt.plot(X_test, y_pred, color='orange', linewidth = 3, label='Predicted Regression Line')
plt.xlabel('BMI')
plt.ylabel('1 year Progression')
plt.title('Simple Linear Regression on Diabetes Dataset')
plt.legend()
plt.show()
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
