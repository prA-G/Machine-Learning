## Example 1

```python
# Reading the dataset
# Here we are loading the data from scikit learn
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
diabetes
```

```python
#Problem Statement: Predicting disease progression one year after baseline provided 10 features.
```

Now since we have target variables, feature names we can build dataFrame

```python
data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
data.head()
```

```python
data['target'] = diabetes.target
```

```python
data
```

### Data Cleaning

Seperating Data into X and Y

```python
X = data[['bmi']] # one independent variable
y = data['target'] # dependent variable
```

```python
X, y
```

### Train-Test-Split

Split array or matrices in random data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

random_state = 1 when fixed helps us to reproduce the same result again and again for 1 data

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Model Training

**Linear Regression** : The following are a set of methods intended for regressions in which the target value is expected to be a linear combination of the features.

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
X_train
```

```python
y_train
```

Training of Model

```python
model.fit(X_train, y_train)
```

At the backend it will use gradient descent to give optimal coefficients

```python
model.coef_  #attribute of the model
```

```python
model.intercept_  #attribute of the model
```

```python
y_pred = model.predict(X_test)  #predicting for a new value
```

```python
y_pred
```

### Visualization of result

```python
plt.scatter(X_test, y_test, color='purple', label='Actual Data')
plt.plot(X_test, y_pred, color='orange', linewidth = 3, label='Predicted Regression Line')
plt.xlabel('BMI')
plt.ylabel('1 year Progression')
plt.title('Simple Linear Regression on Diabetes Dataset')
plt.legend()
plt.show()
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
df = pd.read_csv("height-weight.csv")
```

```python
df.head()
```

## EDA on Data set

```python
plt.scatter(df.Weight, df.Height)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

Problem Statement : **I want to predict height based on weight**

### Divide data into X and y

```python
X = df[['Weight']]  #independent variable
y = df['Height']    #dependent variable
```

```python
X, y
```

### Train-test-split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=1)
```

```python
[X_train, X_test, y_train, y_test]
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Scaling (optional)

>Optional because it doesn't effect distribution of data

Standardization >> fit_transform on train data & transform on test data

**Data Leakage** : If you use fit on test data it will called Data Leakage. It must be avoided.

```python
from sklearn.preprocessing import StandardScaler
```

>Most of the model in ML, Standardscaler is used to deals with normal distributions

```python
scaler = StandardScaler()
scaler
```

```python
X_train
```

- Scaling should be done always after train-test-split
- Traget variable should not be scaled
- Avoid scaling categorical features
- Fit-transform must be used with train data

```python
X_train = scaler.fit_transform(X_train)  #fit_transform on train data
X_test = scaler.transform(X_test)      #transform on test data
```

```python
X_train
```

```python
X_test
```

```python
plt.scatter(X_train, y_train)
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
```

- **Linear Regression Parameter**
    - fit-intercept >> The best fit line will have intercept >> by default it is True
    - copy_X >> Copy the original X_train and then build the model >> It will not modify the original data
    - n_jobs >> The number of jobs used for computations >> -1 means using all the processor
    - positive >> when set to True, all coefficient forced to be positive

- fit >> train the model
- predict >> test the model


- After building the model, you have attributes of the model
    - coef_
    - intercept

```python
regression = LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train, y_train)
```

```python
regression.coef_  #slope of the line
```

```python
regression.intercept_  #y-intercept
```

```python
print("Slope or coefficient of model is : ", regression.coef_)
print("The intercept of the model is : ", regression.intercept_)
```

### Prediction on Train Data

- *Interpretation of Coefficient* >> one unit increase in weight on an avg the height will increase by 17.73 unit
    - `predicted height(y) on X_train = 160 + 17.73 * (X_train)`

```python
regression.predict(X_train)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regression.predict(X_train), color='red')
```

Prediction on Test data

```python
y_pred_test = regression.predict(X_test)
```

```python
y_pred_test, y_test
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred_test, color='brown')
```

### Performance Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred_test)
```

```python
mean_squared_error(y_test, y_pred_test)
```

```python
rsq = r2_score(y_test, y_pred_test)
rsq
```

```python
X_test.shape[0]
```

**Adjusted R Square**

- `[1-(1-rsquare)*n-1/(n-k-1)]`
- n = no. of observation
- k = no of predicted variable

Adj R square always lesser than R sqr

```python
1-(1-rsq)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

- **Assumptions**
    - *X-y should have linear relationship*
    - *Observations should have no relation*
    - *Error should be normally distributed*
    - *Error should have constant variation*
    - *Features among themself should be least related*

```python
plt.scatter(y_test, y_pred_test)
# to check if y_test and y_pred_test are linearly related or aligned
```

```python
error = y_test - y_pred_test
error
```
