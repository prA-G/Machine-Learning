# *Cross Validation , Its types & Hyperparamater Training*

*Scenario* : You trained a model, got 80% or 90% of r2. Now you go and evaluate your model on 1st data. 
- **What if the performances is very bad on trained data?**
    - If in such case you again go back, do feature enginnering, retrain the model, then it will pass the test data information. This will create **Data Leakage**.

## Validation data

It is used to test the model while training itself.

- **But why?**
    - To prevent data leakage
    - Confidence in models accuracy
    - In mpg data, if you don't fix random state, each time we find different rsquare. Now to be confident, each experiments can have slight changes in data. But at the overall level, the avg rsquare/accuracy will gain more confidence in the model.

## Cross Validation

Experimenting with different arrangements of same data to build different models on same algorithm.

- Train-test-split is divided
- Model will be generated and rebust

## Types of Cross Validation

1. **Leave one out CV (LOOC)**
    - If training 100 dps,
        - Exp 1 : 1st data for validation and rest 99 dps for training.
        - Exp 2 : 2nd data for validation and rest other 99 dps for training.
        - and so on Exp 100 : 100th data for validation and rest other 99 for training.
    - Disadvantage:
        - Time complexity is huge for training big data
        - Model overfitting >> Training >> High accuracy
2. **Leave p out CV (LPOC)**
    - Here p > 1. If p = 5 then
        - Exp 1 : 1st 5 data for validation and rest 95 dps for training.
        - Exp 2 : 2nd 5 data for validation and rest other 95 dps for training.
        - and so on Exp 100 : Last 5 data for validation and rest other 95 for training.
3. **K-fold Cross Validation**
    - k = No. of groups you want to divide the data
    - say k = 5 and dps = 100
    - Validation dp = 20 and Training dps = 80
        - Exp 1 : 1st 20 data for validation and rest 80 dps for training.
        - Exp 2 : 2nd 20 data for validation and rest other 80 dps for training.
        - and so on Exp 100 : Last 20 data for validation and rest other 80 for training.
4. **Stratified K-fold CV**
    - For classification problems or imbalance data
    - say k = 5, dps = 100, class = [0, 1]
        - Exp 1 : 1st 20 data for validation where 10 dps for class 0 and other 10 for class 1 and rest 80 dps for training.
        - Exp 2 : 2nd 20 data for validation where 10 dps for class 0 and other 10 for class 1 and rest other 80 dps for training.
        - and so on Exp 100 : Last 20 data for validation where 10 dps for class 0 and other 10 for class 1 and rest other 80 for training.
    - In each validation, there will be equal representation of each class.
