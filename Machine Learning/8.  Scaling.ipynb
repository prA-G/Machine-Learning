# Feature Scaling
- It has three methods
1. Standardization , where mu = 0 & sigma = 1
2. Normlaization (Minmax Scaling) , where data comes between 0 & 1
3. Unit Vector , data = 1

```python
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.distplot(df["total_bill"])
```

```python
mean = np.mean(df["total_bill"])
std = np.std(df["total_bill"])
```

```python
std_data = []
for i in list(df["total_bill"]):
    zscore = (i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

Feature Scaling is optional, it doesn't change the distribution of data

```python
sns.distplot(std_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler
```

```python
scaler.transform(df[['total_bill']])
```

```python
scaler.fit_transform(df[["total_bill", "tip"]])
```

```python
pd.DataFrame(scaler.fit_transform(df[["total_bill", "tip"]]), columns = ['total_bill', 'tip'])
```

```python
scaler.transform([[13,12]])
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
```

```python
min_max.fit_transform(df[['total_bill', 'tip']])
```

```python
min_max.transform([[10, 32]])
```
