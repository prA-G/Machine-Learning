# Data Interpolation :

Process of estimating unknown values from the known values.

## Three types:
- Linear interpolation
- Cubic Interpolation
- Polynomial Interpolation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 3, 5, 7, 9])
```

```python
plt.scatter(x, y)
```

```python
# Linear interpolation >> SHape of data is linear in nature

x_new = np.linspace(1, 5, 10)  # created new x value
y_interp = np.interp(x_new, x, y)
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

```python
# Cubic Interpolation

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x, y, kind = 'cubic')
f
```

```python
x_new = np.linspace(1, 5, 10)
y_interp = f(x_new)
```

```python
plt.scatter(x_new, y_interp)
```

```python
# Polynomial Interpolation
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,1,25])

p = np.polyfit(x, y, 2)
```

```python
x_new = np.linspace(x, y, 10)
y_interp = np.polyval(p, x_new)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x_new, y_interp)
```
