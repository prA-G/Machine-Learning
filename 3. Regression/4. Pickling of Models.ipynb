# Pickle

Allows you to store the python object on the disk

- Python Pickle module is used for serializing and de-serializing a Python object structure. Any object in Python can be pickled so that it can be saved on disk.

### **Serialization**

The procecss of converting a data structure or object into form that can be easily stored, transmitted, or perished. During serialization, the objects state is converted into a stream of byte, or other formats that can be written to disk, sent over a network.

### **Deserialization** 

The process of reconstructing a serialized object back into its original form. During deserialization, the serialized data is read and converted back into a data structure or object that matches the original.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
data
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
X = df.iloc[:,:-1]  #independent features
y = df.iloc[:,-1]   #dependent feature
```

```python
X, y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import pickle
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
pickle.dump(model, open("model.pkl", "wb"))
```

Python objects or model, its attributes and methods are converted into bytes stream

```python
model = pickle.load(open("model.pkl", "rb"))
```

```python
model
```

```python
model.predict(X_test)
```
